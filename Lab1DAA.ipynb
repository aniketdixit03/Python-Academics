```python
import timeit
import matplotlib.pyplot as plt

# Iterative Algorithm
def iterative_sum(N):
    sum = 0
    for i in range(1, N + 1):
        sum += i
    return sum

# Recursive Algorithm
def recursive_sum(N):
    if N == 1:
        return 1
    else:
        return N + recursive_sum(N - 1)

# Measure execution time for both algorithms
N_values = [10, 50, 100, 150, 200, 500, 1000]  # Change this list as needed
iterative_times = []
recursive_times = []

for N in N_values:
    iterative_time = timeit.timeit(lambda: iterative_sum(N), number=1000)
    recursive_time = timeit.timeit(lambda: recursive_sum(N), number=1000)
    iterative_times.append(iterative_time)
    recursive_times.append(recursive_time)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(N_values, iterative_times, label="Iterative")
plt.plot(N_values, recursive_times, label="Recursive")
plt.xlabel("N")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time vs. N")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
import timeit
import matplotlib.pyplot as plt

# Generate a random array of 10,000 elements in the range 1 to 1000
array = [random.randint(1, 1000) for _ in range(10000)]

# Linear Search
def linear_search(arr, key):
    for i, element in enumerate(arr):
        if element == key:
            return i
    return -1

# Binary Search
def binary_search(arr, key):
    left, right = 0, len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            left = mid + 1
        else:
            right = mid - 1
    return -1

# Measure execution time for both search algorithms
search_keys = [random.randint(1, 1000) for _ in range(5)]  # 5 different search keys
linear_times = []
binary_times = []

for key in search_keys:
    linear_time = timeit.timeit(lambda: linear_search(array, key), number=1000)
    binary_time = timeit.timeit(lambda: binary_search(array, key), number=1000)
    linear_times.append(linear_time)
    binary_times.append(binary_time)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.plot(search_keys, linear_times, marker='o', label="Linear Search")
plt.plot(search_keys, binary_times, marker='o', label="Binary Search")
plt.xlabel("Search Key")
plt.ylabel("Execution Time (seconds)")
plt.title("Execution Time for Linear and Binary Searches")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def string_to_integer(s):
    # Base case: if the string is empty, return 0
    if not s:
        return 0
    
    # Convert the first character of the string to an integer
    first_digit = ord(s[0]) - ord('0')
    
    # Recursively convert the rest of the string and add it to the first_digit
    remaining_digits = string_to_integer(s[1:])
    
    # Calculate the final integer value
    result = first_digit * (10 ** (len(s) - 1)) + remaining_digits
    
    return result

# Example usage:
input_string = "13531"
result = string_to_integer(input_string)
print(result)  # Output: 13531
```
