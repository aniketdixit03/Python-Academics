```python
import numpy as np
import matplotlib.pyplot as plt

# Step activation function
def step_function(x):
    return 1 if x >= 0 else 0

# AND gate training data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 0, 0, 1])

# Initial weights and bias
W = np.array([0.2, -0.75])
bias = 10
learning_rate = 0.05

# Lists to store error values and epochs
error_values = []
epochs = 0

# Training loop
while True:
    error = 0
    for i in range(len(X)):
        # Calculate the weighted sum
        weighted_sum = np.dot(X[i], W) + bias
        # Apply the step activation function
        prediction = step_function(weighted_sum)
        # Calculate the error
        error += (Y[i] - prediction) ** 2
        # Update the weights and bias
        delta_w = learning_rate * (Y[i] - prediction) * X[i]
        delta_bias = learning_rate * (Y[i] - prediction)
        W += delta_w
        bias += delta_bias
    # Increment the epoch count
    epochs += 1
    # Append the error for this epoch
    error_values.append(error)
    # Check for convergence (if error is 0 for all training samples)
    if error == 0:
        break

# Plotting the error values over epochs
plt.plot(range(epochs), error_values)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Error Convergence')
plt.show()

# Print the final weights and bias
print("Final Weights after Convergence:", W)
print("Final Bias after Convergence:", bias)
print("Number of Epochs:", epochs)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Step activation function
def step_function(x):
    return 1 if x >= 0 else 0

# OR gate training data
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([0, 1, 1, 1])  # OR gate truth table

# Initial weights and bias
W = np.array([10, 0.2])
bias = -0.75
learning_rate = 0.05

# Lists to store error values and epochs
error_values = []
epochs = 0

# Maximum number of iterations
max_iterations = 1000

# Training loop
while True:
    error = 0
    for i in range(len(X)):
        # Calculate the weighted sum
        weighted_sum = np.dot(X[i], W) + bias
        # Apply the step activation function
        prediction = step_function(weighted_sum)
        # Calculate the error
        error += (Y[i] - prediction) ** 2
        # Update the weights
        delta_w = learning_rate * (Y[i] - prediction) * X[i]
        delta_bias = learning_rate * (Y[i] - prediction)
        W += delta_w
        bias += delta_bias
    # Increment the epoch count
    epochs += 1
    # Append the error for this epoch
    error_values.append(error)
    # Check for convergence or max iterations
    if error <= 0.002 or epochs >= max_iterations:
        break

# Plotting the error values over epochs
plt.plot(range(epochs), error_values)
plt.xlabel('Epochs')
plt.ylabel('Error')
plt.title('Error Convergence')
plt.show()

# Print the final weights and bias
print("Final Weights after Convergence:", W)
print("Final Bias after Convergence:", bias)
print("Number of Epochs:", epochs)
```
